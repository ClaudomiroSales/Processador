```python
def and_gate(input1, input2):
    if input1 == 1 and input2 == 1:
        return 1
    else:
        return 0
    
def not_gate(input):
    if input == 1:
        return 0
    else:
        return 1

print(and_gate(0, 0))
print(and_gate(0, 1))
print(and_gate(1, 0))
print(and_gate(1, 1))

print(not_gate(0))
print(not_gate(1))

#Teste as portas OR e XOR como eu testei a porta AND
def or_gate(input1, input2):
    if input1 == 1 or input2 == 1:
        return 1
    else:
        return 0
    
def xor_gate(input1, input2):
    if input1 != input2:
        return 1
    else:
        return 0
```
